```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as pl
```

```python
def GetHydrusData(mypath,skiprows):
    WB=pd.read_csv('%s/T_Level.out'%mypath,skiprows=skiprows,skipfooter=1,delim_whitespace=True,index_col=0)
    WB.index.name=None
    Qin=-WB['vTop'].values*1000
    Qout=-WB['vBot'].values*1000
    S=WB['Volume'].values*1000
    start='1999-10-1'
    WB['t']=WB.index
    t=pd.to_datetime(start)+WB['t'].apply(pd.Timedelta, unit='D')
    dt=np.hstack([0,np.diff(WB.index)])
    
    df=pd.DataFrame(index=t)
    df['dt']=dt
    df['Qin']=Qin
    df['Qout']=Qout
    df['S']=S
    df['dS']=np.hstack([0,np.diff(S)])
    df['cQ']=np.cumsum((Qin-Qout)*dt)
    return df
```

```python
hyd=GetHydrusData('InfiltrationProblem',[0,1,2,3,4,5,7])
hyd2=GetHydrusData('gfor/input_output',[0,1,2,3,4,6])
```

```python
# pl.plot(ire['dS'].iloc[:300].cumsum(),'r')
# pl.plot(ire['S'].iloc[:300]-ire['S'].iloc[0])

# i=0 #1000
# j=3650 #1100
pl.figure(figsize=(8,6))
pl.subplot(2,1,1)
pl.plot(hyd['dS']-hyd['cQ'].diff(),label='Hydrus 1D',alpha=0.5)
pl.plot(hyd2['dS']-hyd2['cQ'].diff(),label='Gfortran',alpha=0.5)
pl.gca().set_xticklabels('')
# pl.legend(fontsize=13)
pl.grid()
pl.ylabel('Timestep water \nbalance error (mm)',fontsize=13)
pl.subplot(2,1,2)
pl.plot(hyd['dS'].cumsum()-hyd['cQ'],label='Hydrus 1D',alpha=0.5)
pl.plot(hyd2['dS'].cumsum()-hyd2['cQ'],label='Gfortran',alpha=0.5)
# pl.plot(ire['dS'].cumsum()-ire['cQ'],label='Our model, rtol=1E-6')
# pl.plot(ire2['dS'].cumsum()-ire2['cQ'],label='Our model, rtol=1E-7')
pl.legend(fontsize=13)
pl.grid()
pl.ylabel('Cumulated water \nbalance error (mm)',fontsize=13)

pl.subplots_adjust(hspace=0.03)
pl.show()
# pl.plot(ire['dQ'].cumsum().iloc[:300])
# pl.plot(hyd['S'].iloc[:300]-hyd['S'].iloc[0])
```
